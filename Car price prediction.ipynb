```python
car.fueltype.replace(regex=("gas""0","diesel""1"),inplace=True)
car.aspiration.replace(regex=("std""0","turbo""1"),inplace=True)
car.carbody.replace(regex=("sedan""0","hatchback""1","wagon""2","hardtop""3","convertible""4"),inplace=True)
```

```python
#countplot
fig, ax = plt.subplots(2,2,figsize=(16,10)) 
sns.countplot(x='fueltype', data=car, ax=ax[0][0]) 
sns.countplot(x='aspiration', data=car, ax=ax[0][1])
sns.countplot(x= 'carbody', data=car, ax=ax[1][0])
sns.countplot(x= 'price', data=car, ax=ax[1][1])
```

```python
car.corr()
```

```python
#This is a visual representation of our correlation

plt.figure(figsize = (10,10)) 
correlation_matrix = car.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
